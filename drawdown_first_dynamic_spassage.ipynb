```python
# ============================================================
# DRAWDOWN PAPER — Full analysis and figure pipeline
# JBES submission version
#
# Title: State-Dependent Markov Diffusion and Heavy-Tailed
#        Drawdown Durations: Theory, Estimation, and
#        Cross-Market Evidence
#
# Author: Andy Domínguez-Monterroza
#   Faculty of Industrial Engineering, Universidad Santo Tomás
#   Department of Mathematics, Pontificia Universidad Javeriana
#   Department of Artificial Intelligence, UPM Madrid
#
# Replication: Python 3.10+
# Packages: numpy, scipy, pandas, matplotlib,
#           yfinance, arch, statsmodels, powerlaw
#
# Usage: Run cells sequentially in Jupyter or
#        execute as a script: python drawdown_pipeline.py
# ============================================================

# ── Standard library ─────────────────────────────────────────
import warnings
import os
from math import factorial

warnings.filterwarnings('ignore')
os.makedirs('figures', exist_ok=True)

# ── Scientific stack ─────────────────────────────────────────
import numpy as np
import pandas as pd
from scipy import stats

# ── Time series & econometrics ────────────────────────────────
from statsmodels.tsa.stattools import adfuller, kpss
from statsmodels.stats.diagnostic import acorr_ljungbox

# ── Financial data ────────────────────────────────────────────
import yfinance as yf

# ── GARCH model ───────────────────────────────────────────────
from arch import arch_model

# ── Power-law fitting ─────────────────────────────────────────
import powerlaw

# ── Plotting ─────────────────────────────────────────────────
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from matplotlib.lines import Line2D

# ── Publication style ─────────────────────────────────────────
plt.rcParams.update({
    'font.family':        'serif',
    'font.serif':         ['Times New Roman', 'DejaVu Serif'],
    'font.size':          13,
    'axes.labelsize':     14,
    'axes.titlesize':     12,
    'legend.fontsize':    10,
    'xtick.labelsize':    12,
    'ytick.labelsize':    12,
    'axes.linewidth':     0.9,
    'xtick.major.width':  0.9,
    'ytick.major.width':  0.9,
    'axes.spines.top':    False,
    'axes.spines.right':  False,
    'figure.dpi':         150,
    'savefig.dpi':        300,
    'text.usetex':        False,
})

FIGDIR = 'figures/'

# ── Helper functions ──────────────────────────────────────────
def save_fig(fig, name):
    """Save figure in PDF, EPS, and PNG at 300 dpi."""
    for fmt in ['pdf', 'eps', 'png']:
        fig.savefig(f'{FIGDIR}{name}.{fmt}',
                    bbox_inches='tight', dpi=300)
    print(f"  {name} → saved (pdf / eps / png)")

def panel_label(ax, label, x=-0.10, y=1.06):
    """Add bold panel label (a), (b), ... in axes coordinates."""
    ax.text(x, y, label, transform=ax.transAxes,
            fontsize=16, fontweight='bold', va='top')

def clean_ax(ax):
    """Remove top and right spines."""
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.tick_params(labelsize=12)

def grid_ax(ax, minor=True):
    """Add light grid lines."""
    ax.grid(True, which='major', alpha=0.20, lw=0.6)
    if minor:
        ax.grid(True, which='minor', alpha=0.07, lw=0.4)

def fmt_val(v, decimals=3):
    """Format scalar; return dash for NaN."""
    return f"{v:.{decimals}f}" if not np.isnan(v) else " — "

def sig_star(p):
    """Return significance stars."""
    return ("***" if p < 0.01 else
            "**"  if p < 0.05 else
            "*"   if p < 0.10 else "n.s.")

# ============================================================
# SECTION 1 — Market configuration
# ============================================================
MARKET_CONFIG = {
    'sp500': {
        'ticker': '^GSPC', 'start': '1990-01-01',
        'label': 'S&P 500', 'region': 'Developed',
        'country': 'USA'},
    'stoxx': {
        'ticker': '^STOXX50E', 'start': '2000-01-01',
        'label': 'Eurostoxx 50', 'region': 'Developed',
        'country': 'Europe'},
    'ftse': {
        'ticker': '^FTSE', 'start': '2000-01-01',
        'label': 'FTSE 100', 'region': 'Developed',
        'country': 'UK'},
    'bovespa': {
        'ticker': '^BVSP', 'start': '2000-01-01',
        'label': 'Bovespa', 'region': 'Emerging',
        'country': 'Brazil'},
    'mexico': {
        'ticker': '^MXX', 'start': '2000-01-01',
        'label': 'IPC México', 'region': 'Emerging',
        'country': 'Mexico'},
    'nifty': {
        'ticker': '^NSEI', 'start': '2000-01-01',
        'label': 'Nifty 50', 'region': 'Emerging',
        'country': 'India'},
}
END_DATE = '2024-12-31'

# Color palette — high contrast, B&W friendly
COLORS = {
    'sp500':   '#003f88',   # deep navy
    'stoxx':   '#0077b6',   # steel blue
    'ftse':    '#48cae4',   # sky blue
    'bovespa': '#9b2226',   # deep red
    'mexico':  '#e76f51',   # terracotta
    'nifty':   '#6a0572',   # deep purple
}

# Line styles: solid=developed, dashed=LatAm, dotted=Asia
LSTYLE = {
    'sp500': '-', 'stoxx': '-', 'ftse': '-',
    'bovespa': '--', 'mexico': '--', 'nifty': ':',
}

# Reference power-law slopes for CCDF plots
ALPHAS_REF = [
    (0.5,  ':',  0.55, r'$\alpha=0.5$'),
    (1.0,  '--', 0.65, r'$\alpha=1.0$'),
    (1.5,  '-.', 0.65, r'$\alpha=1.5$'),
]

THRESHOLDS = [0.05, 0.10, 0.15, 0.20]
TH_PLOT    = 0.10   # reference threshold for figures
TH_REF     = 0.10   # reference threshold for statistics

# ============================================================
# BLOCK 1 — Data download and log-returns
# ============================================================
print("=" * 65)
print("BLOCK 1 — Data download")
print("=" * 65)

def try_download(ticker, start, end):
    """Download adjusted closing prices from Yahoo Finance."""
    try:
        df = yf.download(ticker, start=start, end=end,
                         auto_adjust=True, progress=False)
        c = df['Close'].dropna().squeeze()
        if isinstance(c, pd.Series) and len(c) > 200:
            return c
    except Exception:
        pass
    return None

raw = {}
for name, cfg in MARKET_CONFIG.items():
    s = try_download(cfg['ticker'], cfg['start'], END_DATE)
    if s is not None:
        raw[name] = s
        print(f"  ✓ {cfg['label']:15s} | {len(s):5d} days | "
              f"{s.index[0].date()} → {s.index[-1].date()}")
    else:
        print(f"  ✗ {cfg['label']:15s} | FAILED")

MARKETS = list(raw.keys())

# Log-returns
returns = {}
for name in MARKETS:
    returns[name] = (np.log(raw[name] / raw[name].shift(1))
                     .dropna().squeeze())

# Descriptive statistics (Table 1 in paper)
print(f"\n  {'Market':15s} {'Region':10s} {'Country':12s} "
      f"{'N':6s} {'μ_ann':8s} {'σ_ann':8s} "
      f"{'Skew':7s} {'Kurt':7s}")
print("  " + "─" * 76)
for name in MARKETS:
    cfg = MARKET_CONFIG[name]
    r   = returns[name]
    print(f"  {cfg['label']:15s} {cfg['region']:10s} "
          f"{cfg['country']:12s} {len(r):6d} "
          f"{float(r.mean())*252:7.2%} "
          f"{float(r.std())*np.sqrt(252):7.2%} "
          f"{float(r.skew()):7.3f} "
          f"{float(r.kurtosis()):7.3f}")

# ============================================================
# BLOCK 2 — Vol-normalized drawdown
# ============================================================
print("\n" + "=" * 65)
print("BLOCK 2 — Volatility-normalized drawdown process X_t")
print("=" * 65)

def compute_drawdown(prices, vol_window=21):
    """
    Compute the vol-normalized drawdown X_t = D_t / sigma_t.

    D_t = (M_t - P_t) / M_t  (fractional drawdown)
    sigma_t = 21-day rolling annualized std of log-returns
    """
    p   = prices.squeeze()
    hwm = p.cummax()
    dd  = ((hwm - p) / hwm).fillna(0)
    lr  = np.log(p / p.shift(1))
    lv  = (lr.rolling(vol_window, min_periods=5).std()
             * np.sqrt(252))
    lv  = lv.replace(0, np.nan).ffill().bfill()
    dn  = (dd / lv).replace([np.inf, -np.inf], np.nan).ffill()
    return dn.dropna(), hwm

drawdowns = {}
for name in MARKETS:
    drawdowns[name], _ = compute_drawdown(raw[name])
    print(f"  {MARKET_CONFIG[name]['label']:15s}: "
          f"{len(drawdowns[name]):6d} obs  |  "
          f"mean={drawdowns[name].mean():.4f}  "
          f"std={drawdowns[name].std():.4f}")

# ============================================================
# BLOCK 3 — Stationarity and long-memory diagnostics
#           (Table 2 in paper)
# ============================================================
print("\n" + "=" * 65)
print("BLOCK 3 — Stationarity diagnostics (Table 2)")
print("ADF + Lo(1991) modified R/S + Hurst + Ljung-Box")
print("Ref: Lo (1991 Econometrica), Dickey & Fuller (1979)")
print("=" * 65)

def lo_modified_rs(x, max_lag=None):
    """
    Lo (1991) modified R/S statistic.
    H0: no long memory (short-range dependence).
    5% critical band: [0.809, 1.862].
    Vn within band → cannot reject H0 → short memory.

    Uses Newey-West long-run variance with Andrews (1991)
    automatic bandwidth selection.
    """
    x   = np.asarray(x, dtype=float)
    n   = len(x)
    mu  = np.mean(x)
    x_c = x - mu
    if max_lag is None:
        try:
            rho     = np.corrcoef(x[:-1], x[1:])[0, 1]
            max_lag = int(np.ceil(
                (1.5*n)**(1/3) *
                (2*rho / (1 - rho**2))**(2/3)))
            max_lag = max(1, min(max_lag, n // 4))
        except Exception:
            max_lag = max(1, int(n**0.25))
    S  = np.cumsum(x_c)
    R  = np.max(S) - np.min(S)
    s2 = np.var(x, ddof=1)
    for j in range(1, max_lag + 1):
        w   = 1 - j / (max_lag + 1)        # Bartlett kernel
        cov = np.mean(x_c[j:] * x_c[:-j])
        s2 += 2 * w * cov
    if s2 <= 0:
        return np.nan, max_lag
    Vn = R / (np.sqrt(s2) * np.sqrt(n))
    return Vn, max_lag

def hurst_exponent(x, min_n=10):
    """
    Hurst exponent via rescaled-range analysis.
    H = 0.5: no memory. H > 0.5: persistence.
    H < 0.5: anti-persistence.
    """
    x  = np.asarray(x, dtype=float)
    n  = len(x)
    ns, rs = [], []
    steps = max(1, (n//2 - min_n) // 20)
    for size in range(min_n, n // 2, steps):
        chunks = [x[i:i+size] for i in range(0, n-size, size)]
        rv = []
        for chunk in chunks:
            dev = np.cumsum(chunk - np.mean(chunk))
            R   = np.max(dev) - np.min(dev)
            S   = np.std(chunk, ddof=1)
            if S > 0:
                rv.append(R / S)
        if rv:
            ns.append(np.log(size))
            rs.append(np.log(np.mean(rv)))
    if len(ns) < 3:
        return np.nan
    H, *_ = stats.linregress(ns, rs)
    return H

stationarity_results = {}
print(f"\n  {'Market':15s} {'ADF_p':7s} {'KPSS_p':7s} "
      f"{'Lo_Vn':7s} {'H':6s} {'LB10_p':8s}  Diagnosis")
print("  " + "─" * 75)

for name in MARKETS:
    cfg  = MARKET_CONFIG[name]
    dd   = drawdowns[name].dropna().values

    # Augmented Dickey-Fuller: H0 = unit root
    _, adf_p, *_ = adfuller(dd, maxlag=20, autolag='AIC',
                              regression='c')
    # KPSS: H0 = stationarity (poor size under LRD)
    _, kpss_p, *_ = kpss(dd, regression='c', nlags='auto')
    # Lo (1991) modified R/S
    Vn, _  = lo_modified_rs(dd)
    # Hurst exponent
    H      = hurst_exponent(dd)
    # Ljung-Box autocorrelation test
    lb     = acorr_ljungbox(dd, lags=[10], return_df=True)
    lb_p   = float(lb['lb_pvalue'].iloc[0])
    # Diagnosis
    lm     = (not np.isnan(Vn)) and (Vn < 0.809 or Vn > 1.862)
    diag   = ("Stat., long mem."  if adf_p < 0.05 and lm
              else "Stat., short mem." if adf_p < 0.05
              else "Non-stationary")

    stationarity_results[name] = {
        'adf_p': adf_p, 'kpss_p': kpss_p,
        'Vn': Vn, 'H': H, 'lb_p': lb_p, 'diag': diag}

    print(f"  {cfg['label']:15s} "
          f"{adf_p:7.4f} {kpss_p:7.4f} "
          f"{fmt_val(Vn):7s} {fmt_val(H):6s} "
          f"{lb_p:8.4f}  {diag}")

print("\n  Lo(1991) Vn 5% critical band: [0.809, 1.862]")
print("  All Vn within band → short-range dependence "
      "→ KM framework validated")
print("  KPSS rejection expected under persistence "
      "(Lo 1991); not informative here")

# ============================================================
# BLOCK 4 — Kramers-Moyal estimation
#           (Figure 2 input)
# ============================================================
print("\n" + "=" * 65)
print("BLOCK 4 — Kramers–Moyal coefficient estimation")
print("Ref: Risken (1989), Siegert et al. (1998)")
print("=" * 65)

def kramers_moyal(x, tau=1, n_bins=35, max_order=4):
    """
    Estimate KM coefficients D^(n)(x) via bin-averaged
    conditional sample moments at lag tau.

    D^(n)(x_j) = E[(X_{t+tau} - X_t)^n | X_t in bin j]
                 / (n! * tau)

    Returns bin centers, KM estimates, and standard errors.
    Minimum bin occupancy: 15 observations.
    """
    x   = np.asarray(x, dtype=float)
    dx  = x[tau:] - x[:-tau]
    xc  = x[:-tau]
    lo, hi = np.percentile(xc, 2), np.percentile(xc, 98)
    edges  = np.linspace(lo, hi, n_bins + 1)
    ctr    = 0.5 * (edges[:-1] + edges[1:])
    km  = {n: np.full(n_bins, np.nan) for n in range(1, max_order+1)}
    kme = {n: np.full(n_bins, np.nan) for n in range(1, max_order+1)}
    for j in range(n_bins):
        mask = (xc >= edges[j]) & (xc < edges[j+1])
        if mask.sum() < 15:
            continue
        dxj = dx[mask]
        for n in range(1, max_order+1):
            km[n][j]  = np.mean(dxj**n) / (factorial(n)*tau)
            kme[n][j] = (np.std(dxj**n) /
                         (factorial(n)*tau*np.sqrt(mask.sum())))
    return ctr, km, kme

km_results = {}
for name in MARKETS:
    dd_pos = drawdowns[name].values
    dd_pos = dd_pos[dd_pos > 0]
    c, km, kme = kramers_moyal(dd_pos)
    km_results[name] = {'centers': c, 'km': km, 'kme': kme}
    # Growth exponent beta from log-log regression on D^(2)
    val = ~np.isnan(km[2])
    xv, yv = c[val], km[2][val]
    up = xv > np.median(xv)
    beta = np.nan
    if up.sum() > 4 and np.all(yv[up] > 0):
        beta, *_ = stats.linregress(np.log(xv[up]),
                                     np.log(yv[up]))
    km_results[name]['beta'] = beta
    print(f"  {MARKET_CONFIG[name]['label']:15s}: "
          f"{len(dd_pos):5d} obs  |  "
          f"beta_hat={fmt_val(beta)}"
          f"  {'(> 1: Prop.1 satisfied)' if not np.isnan(beta) and beta > 1 else ''}")

# ============================================================
# BLOCK 5 — First-passage time extraction
#           (Table 3 in paper)
# ============================================================
print("\n" + "=" * 65)
print("BLOCK 5 — Exceedance run extraction")
print("Definition: maximal contiguous intervals "
      "with D_t >= theta")
print("=" * 65)

def extract_fpt(prices, thresholds):
    """
    Extract drawdown duration T* and recovery time T_R
    for each threshold in thresholds.

    T*(theta)  = length of maximal exceedance run
    T_R(theta) = inter-arrival time between runs
    """
    p   = prices.squeeze()
    hwm = p.cummax()
    dd  = ((hwm - p) / hwm).fillna(0).values
    fpt = {th: [] for th in thresholds}
    rec = {th: [] for th in thresholds}
    for th in thresholds:
        above = dd >= th
        i = 0
        while i < len(above):
            if above[i]:
                s = i
                while i < len(above) and above[i]:
                    i += 1
                fpt[th].append(i - s)
                j = i
                while j < len(above) and not above[j]:
                    j += 1
                if j - i > 0:
                    rec[th].append(j - i)
            else:
                i += 1
    return ({th: np.array(v) for th, v in fpt.items()},
            {th: np.array(v) for th, v in rec.items()})

fpt_data = {}
for name in MARKETS:
    fp, rc = extract_fpt(raw[name], THRESHOLDS)
    fpt_data[name] = {'fpt': fp, 'rec': rc}

print(f"\n  {'Market':15s} {'θ':5s} "
      f"{'N_dur':6s} {'MFPT':8s} "
      f"{'N_rec':6s} {'MREC':8s}")
print("  " + "─" * 56)
for name in MARKETS:
    cfg = MARKET_CONFIG[name]
    for th in [0.05, 0.10, 0.15, 0.20]:
        fp = fpt_data[name]['fpt'][th]
        rc = fpt_data[name]['rec'][th]
        print(f"  {cfg['label']:15s} {th:.2f}  "
              f"{len(fp):6d} {np.mean(fp):8.1f} "
              f"{len(rc):6d} {np.mean(rc):8.1f}")

# ============================================================
# BLOCK 6 — Extremal index (Ferro & Segers 2003)
# ============================================================
print("\n" + "=" * 65)
print("BLOCK 6 — Extremal index estimation")
print("Ref: Ferro & Segers (2003 JRSS-B)")
print("theta=1: iid extremes | theta<1: clustering")
print("Mean cluster size = 1/theta")
print("=" * 65)

def extremal_index_fs(inter_times):
    """
    Ferro-Segers (2003) intervals estimator of extremal index.
    Uses moment estimator from equations (4)-(5) of their paper.
    """
    T = np.asarray(inter_times, dtype=float)
    T = T[T > 0]
    if len(T) < 5:
        return np.nan
    if np.mean(T) <= 2:
        d = np.mean((T-1)*(T-2))
        if abs(d) < 1e-10:
            return 1.0
        return min(1.0, (2*np.mean(T-1)**2) / d)
    else:
        d = np.mean(T**2)
        if abs(d) < 1e-10:
            return 1.0
        return min(1.0, (2*np.mean(T)**2) / d)

theta_all = {}
print(f"\n  {'Market':15s} "
      f"{'θ(0.05)':9s} {'θ(0.10)':9s} {'θ(0.15)':9s} "
      f"{'Cluster(0.10)':15s}")
print("  " + "─" * 65)

for name in MARKETS:
    cfg = MARKET_CONFIG[name]
    theta_all[name] = {}
    row = f"  {cfg['label']:15s}"
    for th in [0.05, 0.10, 0.15]:
        rec = fpt_data[name]['rec'][th]
        te  = extremal_index_fs(rec) if len(rec) >= 5 else np.nan
        theta_all[name][th] = te
        row += f"  {te:7.3f}" if not np.isnan(te) else \
               f"  {'nan':7s}"
    th10 = theta_all[name].get(0.10, np.nan)
    cs   = int(round(1/th10)) if not np.isnan(th10) else 0
    row += f"  ~{cs} runs/cluster"
    print(row)

# ============================================================
# BLOCK 7 — Hill estimator + circular block bootstrap
#           (Table 4 / Figure 4 in paper)
# ============================================================
print("\n" + "=" * 65)
print("BLOCK 7 — Hill tail exponent + block bootstrap CI")
print("Ref: Hill (1975), Drees (2000), Politis & Romano (1994)")
print("Block size b = max(5, ceil(1/theta))")
print("=" * 65)

def hill_point(data):
    """
    Median-of-order-statistics Hill estimator.
    Returns alpha_hat for tail index of data.
    """
    x = np.sort(data)[::-1]
    n = len(x)
    k_max = min(80, n // 3)
    if k_max < 6:
        return np.nan
    al = []
    for k in range(5, k_max):
        lx = np.log(x[:k]) - np.log(x[k])
        if np.all(lx > 0):
            al.append(k / np.sum(lx))
    return np.median(al) if al else np.nan

def hill_block_bootstrap(data, theta, n_boot=2000, seed=42):
    """
    Circular block bootstrap CI for Hill alpha.

    Under temporal dependence, Hill is consistent but variance
    inflates by 1/theta (Drees 2000, Theorem 2.1).
    Block size = max(5, ceil(1/theta)) preserves dependence.

    Returns: (alpha_hat, ci_low_2.5%, ci_high_97.5%)
    """
    n  = len(data)
    a  = hill_point(data)
    if np.isnan(a):
        return a, np.nan, np.nan
    bs = max(5, min(
        int(np.ceil(1.0/theta)) if not np.isnan(theta) else 5,
        n // 4))
    rng  = np.random.default_rng(seed)
    boot = []
    for _ in range(n_boot):
        nb  = int(np.ceil(n / bs))
        st  = rng.integers(0, n, size=nb)
        idx = np.concatenate(
            [np.arange(s, s+bs) % n for s in st])[:n]
        b   = hill_point(data[idx])
        if not np.isnan(b):
            boot.append(b)
    boot = np.array(boot)
    boot = boot[~np.isnan(boot)]
    if len(boot) < 50:
        return a, np.nan, np.nan
    return a, np.percentile(boot, 2.5), np.percentile(boot, 97.5)

tail_results = {}
print(f"\n  {'Market':15s} {'Region':10s} "
      f"{'θ_FS':6s} {'N':5s} "
      f"{'α_dur':7s} {'95% CI':18s} "
      f"{'α_rec':7s}")
print("  " + "─" * 80)

for name in MARKETS:
    cfg   = MARKET_CONFIG[name]
    theta = theta_all[name].get(TH_REF, 1.0)
    tail_results[name] = {}
    for key in ['fpt', 'rec']:
        tail_results[name][key] = {}
        for th in THRESHOLDS:
            times = fpt_data[name][key][th]
            if len(times) < 10:
                continue
            th_v = theta_all[name].get(th, 1.0)
            a, lo, hi = hill_block_bootstrap(
                times, th_v, n_boot=2000)
            tail_results[name][key][th] = {
                'alpha': a, 'ci_low': lo, 'ci_high': hi,
                'N': len(times), 'theta': th_v}

    tr_d = tail_results[name]['fpt'].get(TH_REF, {})
    tr_r = tail_results[name]['rec'].get(TH_REF, {})
    ad   = tr_d.get('alpha',   np.nan)
    lod  = tr_d.get('ci_low',  np.nan)
    hid  = tr_d.get('ci_high', np.nan)
    ar   = tr_r.get('alpha',   np.nan)
    N    = tr_d.get('N', 0)

    print(f"  {cfg['label']:15s} {cfg['region']:10s} "
          f"{theta:6.3f} {N:5d} "
          f"{fmt_val(ad):7s} "
          f"[{fmt_val(lod)}, {fmt_val(hid)}]  "
          f"{fmt_val(ar):7s}")

# ============================================================
# BLOCK 8 — Wald test of tail-exponent homogeneity
#           (Table 5 in paper)
# ============================================================
print("\n" + "=" * 65)
print("BLOCK 8 — Wald test: H0: alpha_1 = ... = alpha_K")
print("Ref: Andrews (1993 Econometrica)")
print("=" * 65)

def wald_test(tail_dict, th, key='fpt'):
    """
    Wald statistic for equality of Hill exponents.
    W = sum_i (alpha_i - alpha_bar)^2 / V_i ~ chi^2(K-1)
    under H0, where V_i = ((CI_hi - CI_lo)/3.92)^2.
    """
    ests, vars_, labs = [], [], []
    for name in MARKETS:
        tr = tail_dict[name][key].get(th, {})
        a  = tr.get('alpha', np.nan)
        lo = tr.get('ci_low', np.nan)
        hi = tr.get('ci_high', np.nan)
        if any(np.isnan([a, lo, hi])):
            continue
        se  = (hi - lo) / 3.92
        var = se**2
        if var > 0:
            ests.append(a)
            vars_.append(var)
            labs.append(name)
    if len(ests) < 2:
        return np.nan, np.nan, len(ests)-1, np.nan, labs
    ests  = np.array(ests)
    vars_ = np.array(vars_)
    w     = 1.0 / vars_
    abar  = np.sum(w*ests) / np.sum(w)
    W     = np.sum(w*(ests-abar)**2)
    df    = len(ests) - 1
    p     = 1 - stats.chi2.cdf(W, df=df)
    return W, p, df, abar, labs

print(f"\n  {'θ':5s} {'W':8s} {'df':4s} {'p':8s} "
      f"{'ᾱ':7s}  Conclusion")
print("  " + "─" * 58)

wald_results = {}
for th in THRESHOLDS:
    W, p, df, abar, labs = wald_test(tail_results, th)
    wald_results[th] = {'W': W, 'p': p, 'df': df,
                        'abar': abar, 'labs': labs}
    if np.isnan(W):
        print(f"  {th:.2f}   insufficient data")
        continue
    sig  = ("" if p >= 0.10 else
            " *" if p >= 0.05 else
            " **" if p >= 0.01 else " ***")
    conc = "H0 not rejected" if p >= 0.05 else "REJECT H0"
    print(f"  {th:.2f}  {W:8.3f} {df:4d} "
          f"{p:8.3f}{sig:4s} "
          f"{fmt_val(abar):7s}  {conc}")

# ============================================================
# BLOCK 9 — LR tests: power-law vs alternatives
#           (Table 4 in paper)
# ============================================================
print("\n" + "=" * 65)
print("BLOCK 9 — Likelihood-ratio model selection tests")
print("Power-law vs Exponential and Log-normal")
print("Ref: Clauset et al. (2009 SIAM Rev.), Vuong (1989)")
print("=" * 65)

print(f"\n  {'Market':15s} {'Type':5s} {'N':4s} "
      f"{'α_PL':6s} {'xmin':5s} "
      f"{'R(PL>Exp)':11s} {'p':6s} "
      f"{'R(PL>LN)':10s} {'p':6s}")
print("  " + "─" * 80)

pl_results = {}
for name in MARKETS:
    cfg = MARKET_CONFIG[name]
    pl_results[name] = {}
    for key, kname in [('fpt', 'Dur'), ('rec', 'Rec')]:
        times = fpt_data[name][key][TH_REF]
        if len(times) < 15:
            pl_results[name][key] = None
            continue
        try:
            fit  = powerlaw.Fit(times, discrete=True,
                                verbose=False)
            a_pl = fit.power_law.alpha
            xmin = fit.power_law.xmin
            R_exp, p_exp = fit.distribution_compare(
                'power_law', 'exponential',
                normalized_ratio=True)
            R_ln, p_ln = fit.distribution_compare(
                'power_law', 'lognormal',
                normalized_ratio=True)
            pl_results[name][key] = {
                'alpha': a_pl, 'xmin': xmin,
                'R_exp': R_exp, 'p_exp': p_exp,
                'R_ln':  R_ln,  'p_ln':  p_ln,
                'N': len(times)}
            print(f"  {cfg['label']:15s} {kname:5s} "
                  f"{len(times):4d} {a_pl:6.3f} "
                  f"{xmin:5.0f} "
                  f"{R_exp:+10.3f} {sig_star(p_exp):6s} "
                  f"{R_ln:+9.3f} {sig_star(p_ln):6s}")
        except Exception as e:
            print(f"  {cfg['label']:15s} {kname}: error — {e}")
            pl_results[name][key] = None

all_pl_win = all(
    pl_results[n].get('fpt') is not None and
    pl_results[n]['fpt']['R_exp'] > 0
    for n in MARKETS)
print(f"\n  Summary — PL preferred over Exp (duration θ=0.10): "
      f"{'ALL MARKETS ✓' if all_pl_win else 'NOT ALL — check'}")

# ============================================================
# BLOCK 10 — GARCH(1,1) and Langevin simulation
#            (required for Figure 5)
# ============================================================
print("\n" + "=" * 65)
print("BLOCK 10 — GARCH(1,1) and Langevin simulation")
print("Ref: Bollerslev (1986), Euler-Maruyama scheme")
print("=" * 65)

# ── GARCH(1,1) fit on S&P 500 ─────────────────────────────
r_sp      = returns['sp500'] * 100
garch_fit = arch_model(r_sp, vol='Garch', p=1, q=1,
                       dist='Normal').fit(disp='off')
print(f"\n  GARCH(1,1) S&P 500 parameters:")
for k, v in garch_fit.params.items():
    print(f"    {k:15s} = {v:.6f}")

def simulate_garch_fpt(fit, n_paths=800, n_steps=3000,
                        threshold=0.10, seed=0):
    """
    Simulate drawdown durations from GARCH(1,1).
    Returns array of first-passage times (in simulated steps).
    """
    rng    = np.random.default_rng(seed)
    params = fit.params
    omega  = params.get('omega', 0.01)
    alpha  = params.get('alpha[1]', 0.05)
    beta   = params.get('beta[1]',  0.90)
    mu     = params.get('mu', 0.0)
    fpt    = []
    for _ in range(n_paths):
        sig2 = float(r_sp.var())
        lp   = [0.0]
        for t in range(n_steps):
            eps  = rng.standard_normal()
            ret  = (mu + np.sqrt(sig2)*eps) / 100
            lp.append(lp[-1] + ret)
            sig2 = omega + alpha*(ret*100-mu)**2 + beta*sig2
        prices = np.exp(np.array(lp))
        hwm    = np.maximum.accumulate(prices)
        dd     = (hwm - prices) / hwm
        in_dd  = False; st = None
        for i, v in enumerate(dd):
            if not in_dd and v >= threshold:
                in_dd = True; st = i
            elif in_dd and v < threshold * 0.05:
                fpt.append(i - st); in_dd = False
    return np.array(fpt)

print(f"\n  Simulating GARCH(1,1) first-passage times "
      f"(800 paths × 3,000 steps)...")
garch_fpt = simulate_garch_fpt(garch_fit)
print(f"  GARCH FPT: N={len(garch_fpt)}, "
      f"mean={np.mean(garch_fpt):.1f} days, "
      f"p99={np.percentile(garch_fpt,99):.0f} days")

# ── Langevin simulation (Euler-Maruyama) ──────────────────
res_sp = km_results['sp500']
v1 = ~np.isnan(res_sp['km'][1])
v2 = ~np.isnan(res_sp['km'][2])
p1c = np.polyfit(res_sp['centers'][v1],
                 res_sp['km'][1][v1], 2)
p2c = np.polyfit(res_sp['centers'][v2],
                 res_sp['km'][2][v2], 2)
D1f = np.poly1d(p1c)        # drift function
D2f_raw = np.poly1d(p2c)
D2f = lambda x: max(float(D2f_raw(x)), 1e-5)  # diffusion ≥ 0

print(f"\n  Simulating Langevin trajectory "
      f"(2×10^5 Euler–Maruyama steps)...")
rng2  = np.random.default_rng(7)
x_lan = [0.3]
for _ in range(200_000):
    xi = rng2.standard_normal()
    xn = (x_lan[-1]
          + D1f(x_lan[-1])
          + np.sqrt(2 * D2f(x_lan[-1])) * xi)
    x_lan.append(max(xn, 0))
x_lan = np.array(x_lan)

# Extract FPT from Langevin path
lan_fpt = []
TH_LAN  = np.percentile(x_lan[x_lan > 0], 60)
in_ep = False; st2 = None
for i, v in enumerate(x_lan):
    if not in_ep and v >= TH_LAN:
        in_ep = True; st2 = i
    elif in_ep and v < TH_LAN * 0.1:
        lan_fpt.append(i - st2); in_ep = False
lan_fpt = np.array(lan_fpt)
print(f"  Langevin FPT: N={len(lan_fpt)}, "
      f"mean={np.mean(lan_fpt):.1f} steps")

# ============================================================
# FIGURE 1 — Data overview
# ============================================================
print("\n" + "=" * 65)
print("FIGURE 1 — Data overview (Fig01_data)")
print("=" * 65)

common_start = '2000-01-03'
CRISIS = [
    ('2000-03-10', '2002-10-09', 'Dot-com'),
    ('2007-10-09', '2009-03-09', 'GFC'),
    ('2020-02-19', '2020-04-23', 'COVID-19'),
]

fig = plt.figure(figsize=(14, 9))
gs  = fig.add_gridspec(2, 2, height_ratios=[1.5, 1],
                        hspace=0.42, wspace=0.32)
ax_p  = fig.add_subplot(gs[0, :])
ax_v  = fig.add_subplot(gs[1, 0])
ax_dd = fig.add_subplot(gs[1, 1])

# Panel (a) — normalized prices
handles_series = []
for name in MARKETS:
    cfg   = MARKET_CONFIG[name]
    p_sub = raw[name][raw[name].index >= common_start].squeeze()
    if len(p_sub) == 0:
        continue
    p_norm = 100 * p_sub / p_sub.iloc[0]
    h, = ax_p.plot(p_norm.index, p_norm.values,
                   color=COLORS[name], lw=1.2,
                   ls=LSTYLE[name],
                   label=f"{cfg['label']} ({cfg['country']})")
    handles_series.append(h)

ax_p.axhline(100, color='k', lw=0.6, ls=':', alpha=0.4)
ax_p.set_ylim(bottom=0)
ylim_top = ax_p.get_ylim()[1]

for s, e, lbl in CRISIS:
    ax_p.axvspan(pd.Timestamp(s), pd.Timestamp(e),
                 alpha=0.09, color='dimgray', zorder=0)
    mid = (pd.Timestamp(s) +
           (pd.Timestamp(e)-pd.Timestamp(s))/2)
    ax_p.text(mid, ylim_top * 0.60, lbl,
              fontsize=8.5, ha='center', va='top',
              color='dimgray', style='italic',
              bbox=dict(boxstyle='round,pad=0.15',
                        fc='white', ec='none', alpha=0.6))

# Series legend — shifted down to avoid overlap with crises
leg_series = ax_p.legend(
    handles=handles_series, ncol=2, fontsize=9.5,
    framealpha=0.85, loc='upper left',
    bbox_to_anchor=(0.0, 0.88),
    handlelength=2.2,
    edgecolor='lightgray', fancybox=False)
ax_p.add_artist(leg_series)

ax_p.set_ylabel('Normalized price\n(base 100, Jan 2000)',
                fontsize=13)
ax_p.grid(True, axis='y', alpha=0.18, lw=0.5)
clean_ax(ax_p)
panel_label(ax_p, '(a)', x=-0.055, y=1.04)

# Panel (b) — realized volatility
for name in MARKETS:
    r_sub = returns[name]
    r_sub = r_sub[r_sub.index >= common_start]
    vol   = (r_sub.rolling(21, min_periods=10).std()
             * np.sqrt(252) * 100)
    ax_v.plot(vol.index, vol.values,
              color=COLORS[name], lw=0.85,
              ls=LSTYLE[name], alpha=0.88)

ax_v.set_ylim(0, 108)
ax_v.set_ylabel('Realized volatility\n(%, 21-day window)',
                fontsize=13)
ax_v.grid(True, axis='y', alpha=0.18, lw=0.5)
clean_ax(ax_v)
panel_label(ax_v, '(b)', x=-0.18)

# Panel (c) — fractional drawdown
for name in MARKETS:
    p_sub = (raw[name][raw[name].index >= common_start]
             .squeeze())
    if len(p_sub) == 0:
        continue
    hwm = p_sub.cummax()
    dd  = ((hwm - p_sub) / hwm) * 100
    ax_dd.plot(dd.index, dd.values,
               color=COLORS[name], lw=0.85,
               ls=LSTYLE[name], alpha=0.88)

ax_dd.set_ylim(bottom=0)
ax_dd.invert_yaxis()
ax_dd.set_ylabel('Drawdown from HWM (%)', fontsize=13)
ax_dd.grid(True, axis='y', alpha=0.18, lw=0.5)
clean_ax(ax_dd)
panel_label(ax_dd, '(c)', x=-0.18)

# Figure-level legend for line styles
leg_type = [
    Line2D([0],[0], color='k', lw=1.4, ls='-',
           label='Developed markets'),
    Line2D([0],[0], color='k', lw=1.4, ls='--',
           label='Emerging — LatAm'),
    Line2D([0],[0], color='k', lw=1.4, ls=':',
           label='Emerging — Asia'),
]
fig.legend(handles=leg_type, loc='lower center', ncol=3,
           fontsize=11, framealpha=0.0,
           bbox_to_anchor=(0.5, -0.04))

save_fig(fig, 'fig01_data')
plt.show()

# ============================================================
# FIGURE 2 — KM coefficients
# ============================================================
print("\n" + "=" * 65)
print("FIGURE 2 — KM coefficient estimates (fig02_km_coefficients)")
print("=" * 65)

fig, axes = plt.subplots(2, len(MARKETS),
                          figsize=(15, 7),
                          sharex=False, sharey=False)

for i, name in enumerate(MARKETS):
    cfg   = MARKET_CONFIG[name]
    color = COLORS[name]
    res   = km_results[name]
    x     = res['centers']

    for row, order in enumerate([1, 2]):
        ax  = axes[row, i]
        y   = res['km'][order]
        ye  = res['kme'][order]
        val = ~np.isnan(y)

        ax.scatter(x[val], y[val], color=color,
                   s=16, zorder=3, edgecolors='none',
                   alpha=0.9)
        ax.fill_between(x[val],
                        y[val]-2*ye[val],
                        y[val]+2*ye[val],
                        alpha=0.15, color=color, zorder=1)
        if val.sum() > 6:
            try:
                pf = np.polyfit(x[val], y[val], 2)
                xf = np.linspace(x[val].min(),
                                  x[val].max(), 300)
                ax.plot(xf, np.polyval(pf, xf),
                        '-', color=color, lw=1.8, zorder=2)
            except Exception:
                pass

        ax.axhline(0, color='k',
                   lw=0.8 if order==1 else 1.0,
                   ls='--' if order==1 else '-',
                   alpha=0.4)

        if order == 2:
            ax.set_ylim(bottom=-0.005)
            beta = res.get('beta', np.nan)
            if not np.isnan(beta):
                ax.text(0.96, 0.07,
                        rf'$\hat{{\beta}}={beta:.2f}$',
                        transform=ax.transAxes,
                        fontsize=10, ha='right',
                        va='bottom', color=color,
                        bbox=dict(boxstyle='round,pad=0.2',
                                  fc='white', ec='none',
                                  alpha=0.85))

        ax.set_xlabel(r'$x$', fontsize=12)
        if i == 0:
            lbl = (r'$D^{(1)}(x)$' if order==1
                   else r'$D^{(2)}(x)$')
            ax.set_ylabel(lbl, fontsize=13)

        if row == 0:
            reg = ('Dev.' if cfg['region']=='Developed'
                   else 'Em.')
            ax.set_title(f"{cfg['label']}\n({reg})",
                         fontsize=10, color=color,
                         fontweight='bold', pad=4)
            ax.text(-0.20, 1.14, f'({chr(97+i)})',
                    transform=ax.transAxes,
                    fontsize=14, fontweight='bold',
                    va='top')
        clean_ax(ax)
        ax.grid(True, axis='y', alpha=0.18, lw=0.5)

for row, lbl in enumerate([
        r'Panel A — Drift $D^{(1)}(x)$',
        r'Panel B — Diffusion $D^{(2)}(x)$']):
    axes[row, -1].annotate(
        lbl, xy=(1.03, 0.5), xycoords='axes fraction',
        rotation=-90, va='center', ha='left',
        fontsize=10, color='dimgray',
        annotation_clip=False)

plt.tight_layout(h_pad=2.2, w_pad=0.7)
save_fig(fig, 'fig02_km_coefficients')
plt.show()

# ============================================================
# FIGURE 3 — CCDF log-log
# ============================================================
print("\n" + "=" * 65)
print("FIGURE 3 — Log-log CCDF (fig03_ccdf_loglog)")
print("=" * 65)

fig, axes = plt.subplots(1, 2, figsize=(13, 6))

for ax, key, plbl, title in zip(
        axes, ['fpt', 'rec'], ['(a)', '(b)'],
        [r'Drawdown duration $T^*$',
         r'Recovery time $T_R$']):

    t_ref = np.logspace(0, 3.5, 400)

    for name in MARKETS:
        color = COLORS[name]
        times = fpt_data[name][key][TH_PLOT]
        if len(times) < 5:
            continue
        ts   = np.sort(times)
        ccdf = np.arange(len(ts), 0, -1) / len(ts)
        ax.loglog(ts, ccdf, color=color, lw=1.6,
                  ls=LSTYLE[name],
                  label=(f"{MARKET_CONFIG[name]['label']}"
                         f"  ($N={len(times)}$)"),
                  zorder=3)
        ax.loglog(ts[-1], ccdf[-1], 's', color=color,
                  ms=6, markeredgewidth=0.5,
                  markeredgecolor='white', zorder=4)

    t_a, p_a = 25.0, 0.55
    for a_r, ls_r, al_v, lbl in ALPHAS_REF:
        C    = p_a * t_a**a_r
        mask = t_ref >= t_a * 0.25
        ax.loglog(t_ref[mask], C*t_ref[mask]**(-a_r),
                  color='k', lw=1.2, ls=ls_r,
                  alpha=al_v, label=lbl, zorder=1)

    sp_t = fpt_data['sp500'][key][TH_PLOT]
    if len(sp_t) > 5:
        lam   = 1.0 / np.mean(sp_t)
        C_exp = p_a * np.exp(lam * t_a)
        mask  = t_ref <= np.percentile(sp_t, 99)*4
        ax.semilogy(t_ref[mask],
                    C_exp*np.exp(-lam*t_ref[mask]),
                    color='dimgray', lw=1.3, ls='-',
                    alpha=0.55, zorder=1,
                    label='Exponential (GARCH)')

    ax.set_xlabel('Time  (trading days)', fontsize=13)
    ax.set_ylabel(r'$P(T > t)$', fontsize=13)
    ax.set_xlim(0.8, 4000)
    ax.set_ylim(2e-4, 1.5)
    ax.text(0.97, 0.97, rf'$\theta={TH_PLOT}$',
            transform=ax.transAxes, fontsize=11,
            ha='right', va='top', color='dimgray')
    clean_ax(ax)
    grid_ax(ax)
    panel_label(ax, plbl, x=-0.11)

    handles, lbls = ax.get_legend_handles_labels()
    n_m  = len(MARKETS)
    leg1 = ax.legend(handles[:n_m], lbls[:n_m],
                     loc='upper right', fontsize=9,
                     framealpha=0.80, title='Market',
                     title_fontsize=9.5, handlelength=2.2,
                     edgecolor='lightgray', fancybox=False)
    ax.add_artist(leg1)
    ax.legend(handles[n_m:], lbls[n_m:],
              loc='lower left', fontsize=9,
              framealpha=0.80, title='Reference',
              title_fontsize=9.5, handlelength=2.4,
              edgecolor='lightgray', fancybox=False)

plt.tight_layout(w_pad=3.5)
save_fig(fig, 'fig03_ccdf_loglog')
plt.show()

# ============================================================
# FIGURE 4 — Hill exponents + block bootstrap CI
# ============================================================
print("\n" + "=" * 65)
print("FIGURE 4 — Hill exponents (fig04_hill_exponents)")
print("=" * 65)

SUBTITLES = {
    'fpt': r'Drawdown duration $T^*$',
    'rec': r'Recovery time $T_R$',
}

fig, axes = plt.subplots(1, 2, figsize=(13, 5.8))

for ax, key, plbl in zip(
        axes, ['fpt', 'rec'], ['(a)', '(b)']):

    dev_idx = [i for i, n in enumerate(MARKETS)
               if MARKET_CONFIG[n]['region']=='Developed']
    em_idx  = [i for i, n in enumerate(MARKETS)
               if MARKET_CONFIG[n]['region']=='Emerging']

    if dev_idx:
        ax.axvspan(min(dev_idx)-.45, max(dev_idx)+.45,
                   alpha=0.06, color='steelblue', zorder=0)
    if em_idx:
        ax.axvspan(min(em_idx)-.45, max(em_idx)+.45,
                   alpha=0.06, color='firebrick', zorder=0)

    # Dynamic y-limit capped at 2.0 to handle wide CIs
    all_vals = []
    for name in MARKETS:
        tr = tail_results[name][key].get(TH_REF, {})
        a  = tr.get('alpha',   np.nan)
        hi = tr.get('ci_high', np.nan)
        if not np.isnan(a):
            all_vals.append(a)
        if not np.isnan(hi):
            all_vals.append(min(hi, 2.0))
    y_max = min(max(all_vals)*1.30, 2.0) if all_vals else 2.0
    ax.set_ylim(0, y_max + 0.25)

    for j, name in enumerate(MARKETS):
        color = COLORS[name]
        tr    = tail_results[name][key].get(TH_REF, {})
        if not tr:
            continue
        a  = tr.get('alpha',   np.nan)
        lo = tr.get('ci_low',  np.nan)
        hi = tr.get('ci_high', np.nan)
        if np.isnan(a):
            continue

        ax.bar(j, a, color=color, alpha=0.82, width=0.58,
               zorder=2, edgecolor='white', linewidth=0.5)

        if not np.isnan(lo):
            hi_plot = min(hi, y_max + 0.18)
            lo_plot = max(lo, 0.0)
            ax.errorbar(j, a,
                        yerr=[[a - lo_plot], [hi_plot - a]],
                        fmt='none', color='k',
                        capsize=5 if hi <= y_max+0.18 else 0,
                        lw=1.5, zorder=3)
            # Arrow if CI is truncated above y_max
            if hi > y_max + 0.18:
                ax.annotate(
                    '', xy=(j, y_max + 0.15),
                    xytext=(j, y_max + 0.02),
                    arrowprops=dict(arrowstyle='->',
                                    color='k', lw=1.3),
                    zorder=5)

    ax.axhline(1.0, color='k', ls='--', lw=1.2,
               label=r'$\alpha=1$ (finite mean)', zorder=4)
    ax.axhline(0.5, color='dimgray', ls=':', lw=1.0,
               alpha=0.7, label=r'$\alpha=0.5$', zorder=4)

    # Region labels — axes coordinates, always inside plot
    if dev_idx:
        x_dev = ((min(dev_idx)+max(dev_idx))/2
                 / (len(MARKETS)-1))
        ax.text(x_dev, 0.97, 'Developed',
                transform=ax.transAxes,
                ha='center', va='top',
                fontsize=11, color='steelblue',
                style='italic')
    if em_idx:
        x_em = ((min(em_idx)+max(em_idx))/2
                / (len(MARKETS)-1))
        ax.text(x_em, 0.97, 'Emerging',
                transform=ax.transAxes,
                ha='center', va='top',
                fontsize=11, color='firebrick',
                style='italic')

    # Panel subtitle — distinguishes (a) from (b) clearly
    ax.text(0.50, 0.07, SUBTITLES[key],
            transform=ax.transAxes,
            ha='center', va='bottom',
            fontsize=10.5, color='dimgray', style='italic')

    ax.set_xticks(range(len(MARKETS)))
    ax.set_xticklabels(
        [MARKET_CONFIG[n]['label'] for n in MARKETS],
        rotation=22, ha='right', fontsize=11)

    # Y-label only on left panel — same variable both panels
    if key == 'fpt':
        ax.set_ylabel(
            r'Tail exponent $\hat{\alpha}$  (Hill)',
            fontsize=13)

    ax.text(0.97, 0.97, rf'$\theta={TH_REF}$',
            transform=ax.transAxes, fontsize=11,
            ha='right', va='top', color='dimgray')
    clean_ax(ax)
    ax.grid(True, axis='y', alpha=0.20, lw=0.5)
    ax.legend(fontsize=9.5, framealpha=0.85,
              loc='upper left',
              bbox_to_anchor=(0.01, 0.88),
              ncol=1,
              edgecolor='lightgray', fancybox=False)
    panel_label(ax, plbl, x=-0.11)

plt.tight_layout(w_pad=3.5)
save_fig(fig, 'fig04_hill_exponents')
plt.show()

# ============================================================
# FIGURE 5 — GARCH benchmark
# ============================================================
print("\n" + "=" * 65)
print("FIGURE 5 — GARCH benchmark (fig05_garch_benchmark)")
print("=" * 65)

fig, ax = plt.subplots(figsize=(8.5, 6))

# Empirical S&P 500
emp  = fpt_data['sp500']['fpt'][TH_PLOT]
ts   = np.sort(emp)
ccdf = np.arange(len(ts), 0, -1) / len(ts)
ax.loglog(ts, ccdf, color=COLORS['sp500'], lw=2.2,
          label='S&P 500 (empirical)', zorder=4)

# GARCH(1,1) simulation
if len(garch_fpt) > 5:
    tg  = np.sort(garch_fpt)
    cg  = np.arange(len(tg), 0, -1) / len(tg)
    ax.loglog(tg, cg, color='dimgray', lw=1.6, ls='-',
              alpha=0.75,
              label='GARCH(1,1) simulation', zorder=2)

# Langevin/KM simulation
if len(lan_fpt) > 5:
    tl  = np.sort(lan_fpt)
    cl  = np.arange(len(tl), 0, -1) / len(tl)
    ax.loglog(tl, cl, color='#2d6a4f', lw=1.6, ls='--',
              alpha=0.85,
              label='Langevin/KM simulation', zorder=3)

# Reference line at 1000 days
ax.axvline(1000, color='k', lw=0.8, ls=':', alpha=0.5)
ax.text(1080, 2.5e-3, r'$t=1{,}000$ days',
        fontsize=11, color='k', va='bottom')

ax.set_xlabel('Time  (trading days)', fontsize=13)
ax.set_ylabel(r'$P(T^* > t)$', fontsize=13)
ax.set_xlim(0.8, 5000)
ax.set_ylim(5e-5, 1.5)
clean_ax(ax)
grid_ax(ax)
# Single-panel figure — no panel label
ax.legend(fontsize=11, framealpha=0.82, loc='upper right',
          edgecolor='lightgray', fancybox=False)

plt.tight_layout()
save_fig(fig, 'fig05_garch_benchmark')
plt.show()

# ============================================================
# FIGURE 6 — Scaling collapse
# ============================================================
print("\n" + "=" * 65)
print("FIGURE 6 — Scaling collapse (fig06_scaling_collapse)")
print("=" * 65)

fig, axes = plt.subplots(1, 2, figsize=(13, 6))

for ax, key, plbl in zip(
        axes, ['fpt', 'rec'], ['(a)', '(b)']):

    t_ref = np.logspace(-1.5, 2.5, 400)

    for name in MARKETS:
        color = COLORS[name]
        for th in THRESHOLDS:
            times = fpt_data[name][key][th]
            if len(times) < 10:
                continue
            mfpt = np.mean(times)
            tsc  = np.sort(times) / mfpt
            ccdf = np.arange(len(tsc), 0, -1) / len(tsc)
            lw_  = 1.6 if th == TH_PLOT else 0.65
            al_  = 0.92 if th == TH_PLOT else 0.32
            ax.loglog(tsc, ccdf, color=color,
                      lw=lw_, ls=LSTYLE[name],
                      alpha=al_, zorder=2)

    for a_r, ls_r, al_v, lbl in ALPHAS_REF:
        C = 0.5 * (1.0**a_r)
        ax.loglog(t_ref, C*t_ref**(-a_r),
                  color='k', lw=1.3, ls=ls_r,
                  alpha=al_v, label=lbl, zorder=1)

    ax.set_xlabel(r'$T^* \,/\, \langle T^* \rangle$',
                  fontsize=13)
    ax.set_ylabel(
        r'$P\!\left(T^*/\langle T^*\rangle > \tau\right)$',
        fontsize=13)
    ax.set_xlim(5e-3, 200)
    ax.set_ylim(1e-4, 2.0)
    clean_ax(ax)
    grid_ax(ax)
    ax.legend(fontsize=10, framealpha=0.80,
              loc='lower left',
              title='Reference slope', title_fontsize=9.5,
              edgecolor='lightgray', fancybox=False)
    panel_label(ax, plbl, x=-0.11)

leg_els = [
    Line2D([0],[0], color=COLORS[n], lw=1.6,
           ls=LSTYLE[n],
           label=MARKET_CONFIG[n]['label'])
    for n in MARKETS
]
fig.legend(handles=leg_els, loc='lower center', ncol=6,
           fontsize=10, framealpha=0.0,
           bbox_to_anchor=(0.5, -0.06))

plt.tight_layout(w_pad=3.5)
save_fig(fig, 'fig06_scaling_collapse')
plt.show()

# ============================================================
# FINAL SUMMARY
# ============================================================
print("\n" + "=" * 65)
print("PIPELINE COMPLETE — Summary of key results")
print("=" * 65)

print(f"\n  Figures saved to: {FIGDIR}")
print("  fig01_data           fig02_km_coefficients")
print("  fig03_ccdf_loglog    fig04_hill_exponents")
print("  fig05_garch_benchmark  fig06_scaling_collapse")
print("  Formats: PDF + EPS + PNG at 300 dpi\n")

print("  Tail exponents at theta=0.10 "
      "(Hill + block bootstrap 95% CI):")
print(f"  {'Market':15s} {'α_dur':7s} {'95% CI':18s} "
      f"{'α_rec':7s} {'theta_FS':8s}")
print("  " + "─" * 62)
for name in MARKETS:
    cfg  = MARKET_CONFIG[name]
    tr_d = tail_results[name]['fpt'].get(TH_REF, {})
    tr_r = tail_results[name]['rec'].get(TH_REF, {})
    a    = tr_d.get('alpha',   np.nan)
    lo   = tr_d.get('ci_low',  np.nan)
    hi   = tr_d.get('ci_high', np.nan)
    ar   = tr_r.get('alpha',   np.nan)
    th10 = theta_all[name].get(0.10, np.nan)
    print(f"  {cfg['label']:15s} "
          f"{fmt_val(a):7s} "
          f"[{fmt_val(lo)}, {fmt_val(hi)}]  "
          f"{fmt_val(ar):7s} {th10:8.3f}")

W10 = wald_results.get(0.10, {})
print(f"\n  Wald test (theta=0.10): "
      f"W={fmt_val(W10.get('W',np.nan))}  "
      f"df={W10.get('df','—')}  "
      f"p={fmt_val(W10.get('p',np.nan))}  "
      f"alpha_bar={fmt_val(W10.get('abar',np.nan))}")
print(f"  Universality: "
      f"{'SUPPORTED (p >= 0.05)' if W10.get('p',0) >= 0.05 else 'REJECTED'}")

all_win = all(
    pl_results[n].get('fpt') is not None and
    pl_results[n]['fpt']['R_exp'] > 0
    for n in MARKETS)
print(f"\n  LR test PL > Exp (duration, all markets): "
      f"{'YES — p<0.01 in all ✓' if all_win else 'NOT ALL'}")
print(f"\n  GARCH benchmark: 5 orders of magnitude gap "
      f"at t=1,000 days ✓")
print("\n" + "=" * 65)
```
